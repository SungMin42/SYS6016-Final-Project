```python
# Reading in the data for 'APPL' stock to np.array
hf = h5py.File('../data/price_level_total_view_2017-01-03.h5', 'r')
# List all datasets
print("Keys: %s" % list(hf.keys()))

data = hf.get('AAPL').value
```

```python
# Reading into pandas dataframe
pddf = pd.DataFrame()
for i in range(len(data.dtype.names)):
    dummy = [row[i] for row in data]
    pddf[data.dtype.names[i]] = dummy
```

```python
# Format of dataframe
pddf
```

```python
# Digging deeper into the levels
pddf['levels'][100]
```

So it is clear that asks come before bids. The levels data is essentially a list of tuples (price, quantity, order count).
Not too sure what order count is.

```python
# Mid Price Calculation
best_ask = np.array([y[0] for y in [x[4] for x in pddf['levels']]])
best_bid = np.array([y[0] for y in [x[5] for x in pddf['levels']]])
mid_price = (best_ask + best_bid) / 2
pddf['mid_price'] = mid_price
```

```python
# Quick Calculation Check
print('Best ask for index 100 is:', best_ask[100])
print('Best bid for index 100 is:', best_bid[100])
print('Mid-price for index 100 is:', mid_price[100])
```

```python
pddf
```

```python
# Group by millisecond
# New grouped will be called pddf1
pddf1 = pddf.iloc[:, ]
pddf1['groupid'] = np.floor((pddf.timestamp - pddf.timestamp[0]) / 1000000).astype(np.int32)
pddf1 = pddf1.groupby('groupid').last()
pddf1
```

```python
# Creating target mid_price by shifting mid_price
pddf1['mid_price_shift'] = pddf1['mid_price'].shift(periods = -1)
pddf1
```

```python
# Doing the same for the ungrouped dataset
pddf['mid_price_shift'] = pddf['mid_price'].shift(periods = -1)
pddf
```

```python
# Subsetting columns
pddf_1 = pddf[['trade_volume', 'mid_price', 'mid_price_shift']]
pddf1_1 = pddf1[['trade_volume', 'mid_price', 'mid_price_shift']]
```

```python
pddf_1
```

```python
# Pickling
pddf_1.to_pickle('../data/price_level_total_view_2017-01-03_AAPL_ungrouped')
pddf1_1.to_pickle('../data/price_level_total_view_2017-01-03_AAPL_grouped')
```
