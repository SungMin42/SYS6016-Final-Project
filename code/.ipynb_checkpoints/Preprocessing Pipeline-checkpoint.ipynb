```python
# Reading in the data for 'APPL' stock to np.array
hf = h5py.File('../data/price_level_total_view_2017-01-03.h5', 'r')
# List all datasets
print("Keys: %s" % list(hf.keys()))

data = hf.get('AAPL').value
```

```python
# Reading into pandas dataframe
pddf = pd.DataFrame()
for i in range(len(data.dtype.names)):
    dummy = [row[i] for row in data]
    pddf[data.dtype.names[i]] = dummy
```

```python
# Format of dataframe
pddf
```

```python
# Digging deeper into the levels
pddf['levels'][100]
```

So it is clear that asks come before bids. The levels data is essentially a list of tuples (price, quantity, order count).
Not too sure what order count is.

```python
# Mid Price Calculation
best_ask = np.array([y[0] for y in [x[4] for x in pddf['levels']]])
best_bid = np.array([y[0] for y in [x[5] for x in pddf['levels']]])
mid_price = (best_ask + best_bid) / 2
pddf['mid_price'] = mid_price
```

```python
# Quick Calculation Check
print('Best ask for index 100 is:', best_ask[100])
print('Best bid for index 100 is:', best_bid[100])
print('Mid-price for index 100 is:', mid_price[100])
```

```python
# cheacking progress thus far
pddf.head(n = 20)
```

```python
# calculating log of mid_price and getting rid of log values of -inf from the data
pddf['mid_price_log'] = np.log(pddf['mid_price'])
pddf = pddf[np.isinf(pddf['mid_price_log']) == False]
```

```python
# check
pddf.head(n = 20)
```

```python
# Group by millisecond
# New grouped will be called pddf1
pddf1 = pddf.iloc[:, ]
pddf1['groupid'] = np.ceil((pddf.timestamp - pddf.iloc[0, :].timestamp) / 1000000).astype(np.int32)
pddf1 = pddf1.groupby('groupid').last()
pddf1
```

```python
# Creating target 
pddf['mid_price_log_shift'] = pddf['mid_price_log'].shift(periods = -1)
pddf1['mid_price_log_shift'] = pddf1['mid_price_log'].shift(periods = -1)

pddf['target'] = np.select([(pddf['mid_price_log_shift'] > pddf['mid_price_log']), (pddf['mid_price_log_shift'] == pddf['mid_price_log']), (pddf['mid_price_log_shift'] < pddf['mid_price_log'])], [2, 1, 0])
pddf1['target'] = np.select([(pddf1['mid_price_log_shift'] > pddf1['mid_price_log']), (pddf1['mid_price_log_shift'] == pddf1['mid_price_log']), (pddf1['mid_price_log_shift'] < pddf1['mid_price_log'])], [2, 1, 0])
```

```python
# checking target creation works as intended
pddf1.groupby('target').count()
```

```python
# Subsetting columns
pddf_1 = pddf[['trade_volume', 'mid_price_log', 'target']]
pddf1_1 = pddf1[['trade_volume', 'mid_price_log', 'target']]
```
